```python
import sys
import os
sys.path.insert(0, os.path.abspath(".."))
import copy

from monte_carlo import MonteCarlo
from experiment import Experiment
from basic_model_config import *
from utils import fill_config
```

```python
#reproduce
import copy

exp2_configs = []
base_configs = [basic_model_config_gsgp]

for baseconfig in base_configs:
    

    for fitness in ['sigmoid_rmse', 'f1_score']:
        if fitness == 'f1_score':
            config = copy.deepcopy(baseconfig)  # Ensure fresh copy
            config = fill_config(config, oversampling=False, fitness_function=fitness, minimization=False)
            exp2_configs.append(config)
        else:
            config = copy.deepcopy(baseconfig)  # Ensure fresh copy
            config = fill_config(config, oversampling=False, fitness_function=fitness, minimization=True)
            exp2_configs.append(config)
        


# Print results to verify
for config in exp2_configs:
    if 'slim_version' in config['config']:
        print(config['config']['slim_version'], 'Fitness_Function', config['config']['fitness_function'], 'p_inflate', config['config']['p_inflate'])
    else:
        print(config['name'], 'Fitness_Function', config['config']['fitness_function'])
```

```python
exp2 = Experiment(
                "experiment_comparison",
                "../../data/",
                exp2_configs,
                n_runs=5, 
                log=True, 
                verbose = True
                )
exp2.run()
```
