```python
import copy
rq2_configs = []
base_configs = [ basic_model_config_slim_plussig1]

for baseconfig in base_configs:

        for inflation_rate in [0.3, 0.95]:
                for ms_upper in [0.5, 0.1]:

                        config = copy.deepcopy(baseconfig)  # Fresh copy for each function
                        config = fill_config(config, oversampling= False, fitness_function='sigmoid_rmse', minimization=True, inflation_rate=inflation_rate)
                        config['config']['ms_upper'] = ms_upper
                        rq2_configs.append(config)

for config in rq2_configs:
        if config['name'] == 'slim':
                print(config['config']['slim_version'], config['config']['ms_upper'], config['config']['p_inflate'])
        else:
                print(config['name'])
```
