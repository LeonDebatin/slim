```python
#https://openml.org/search?type=data&sort=runs&status=active&id=1464
blood = loadarff('../data/data_original/blood.arff')
blood = pd.DataFrame(blood[0])
blood['Class'] = blood['Class'].apply(lambda x: 1 if x == b'2' else 0)
blood.rename(columns={'Class': 'target'}, inplace=True)
basic_information(blood)
blood.to_csv('../data/data_prepared/blood.csv', index=False)
info_df.append(info_dict('blood', blood, []))
blood
```

```python
#https://archive.ics.uci.edu/dataset/252/clima+model+simulation+crashes
with open("../data/data_original/clima.dat", "r") as file:
    data = file.readlines()
    
columns = data[0].strip().split()
values = [list(map(float, row.strip().split())) for row in data[1:]]
clima = pd.DataFrame(values, columns=columns)
clima.rename(columns={'outcome': 'target'}, inplace=True)
#clima['target'] = clima['target'].astype(int).replace({0: 1, 1: 0})
clima.drop(columns=['Study', 'Run'], inplace=True)
clima.to_csv('../data/data_prepared/clima.csv', index=False)
info_df.append(info_dict('clima', clima, []))
basic_information(clima)
clima
```

```python
#https://archive.ics.uci.edu/dataset/244/fertility
with open("../data/data_original/fertility.txt", "r") as file:
    data = file.readlines()
fertility = pd.DataFrame([data[i].strip().split(",") for i in range(len(data))], columns=["season", "age", "child_diseases", "accident", "surgical_intervention", "high_fevers", "alcohol", "smoking", "hr_sittings", "diagnosis"])

fertility.rename(columns={'diagnosis': 'target'}, inplace=True)
fertility['target'] = fertility['target'].apply(lambda x: 0 if x == 'N' else 1)
fertility = fertility.astype(float)
fertility['target'] = fertility['target'].astype(int)
fertility.to_csv('../data/data_prepared/fertility.csv', index=False)
info_df.append(info_dict('fertility', fertility, ['season', 'child_diseases', 'accident', 'surgical_intervention', 'high_fevers', 'alcohol', 'smoking']))
basic_information(fertility)
fertility
```
